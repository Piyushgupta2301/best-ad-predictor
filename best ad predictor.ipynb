```python
import pandas as pd
```

```python
df=pd.read_csv('ads.csv')
```

```python
df.head(10)
```

```python
import math
n=10000
d=10
ad_selected=[]
no_of_selected=[0]*d
sum_of_rewards=[0]*d
total_reward=0

for i in range(0,n):
    ad=0
    max_upbound=0
    for j in range(0,d):
        if (no_of_selected[j]>0):
            avg=sum_of_rewards[j]/no_of_selected[j]
            delta_i=math.sqrt(3/2 * math.log(n+1)/no_of_selected[j])
            upperbound=avg+delta_i
        else:
            upperbound=1e400
        
        if upperbound>max_upbound:
            max_upbound=upperbound
            ad=j
            
    ad_selected.append(ad)
    no_of_selected[ad] +=1
    reward=df.values[i,ad]
    sum_of_rewards[ad]+=reward
    total_reward+=reward
            
```

```python
import matplotlib.pyplot as plt
plt.hist(ad_selected)
```
